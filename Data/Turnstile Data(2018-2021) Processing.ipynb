```python
df18 = df18.rename(columns ={'Exits                                                     ': 'Exits'})
```

```python
df18['Entries']=df18['Entries'].str.replace(',','')
```

```python
df18['Entries']=df18['Entries'].astype(int)
```

```python
df18['Exits']=df18['Exits'].str.replace(',','')
df18['Exits']=df18['Exits'].astype(int)
```

```python
daily2018 = df18.groupby('date').agg(
    sum_exits = pd.NamedAgg(column='Exits', aggfunc=sum),
    sum_entries = pd.NamedAgg(column='Entries', aggfunc=sum),    
).reset_index()
    
    
```

```python
daily2018['loss']=daily2018['sum_exits']-daily2018['sum_entries']
```
